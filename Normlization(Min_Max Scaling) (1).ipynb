# Feature Scaling (Normalization)

# * Min-Max Scaling

* It is a scaling technique in which values are shifted and rescaled so that they end up ranging between  0 and 1 .  It is also known as Min-Max Scaling .

X(new)=(X(i)-min(X))/max(x)-min(X)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("C:\\Users\\ASUS\\OneDrive\\Desktop\\pandas\\loan.csv")
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
# Min-Max Scaling
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
ms.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_min_max"]=ms.transform(dataset[["CoapplicantIncome"]])
```

```python
dataset.head(5)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["CoapplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["CoapplicantIncome_min_max"])

plt.show()
```
