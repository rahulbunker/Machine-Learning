## Convertors

```python
def rename(sex):
    if sex=="Female":
        return "girl"
    else:
        return "boy"
```

```python
rename("Female")
```

```python
df=pd.read_csv("C:\\Users\\ASUS\\OneDrive\\Desktop\\pandas\\tips.csv",converters={"sex":rename})
df
# we pass colm name and function name
```
