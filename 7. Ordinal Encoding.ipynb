```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.DataFrame({"size":['s','m','l','xl','xxl','m','s','l']})
```

```python
dataset
```

```python
  data_order = [['s','m','l','xl','xxl']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = data_order)
oe.fit(dataset[["size"]])
# if hame data abcd se according chahiye hota to categories = auto dalte
```

```python
oe.transform(dataset[["size"]])
```

```python
dataset["size_en"] = oe.transform(dataset[["size"]])
dataset
```

```python
data_order_2 = {'s':00 , "m" :11 , "l" : 2 ,"xl" : 3 , "xxl" : 4}
```

```python
dataset["size"].map(data_order_2)
```

```python
dataset["size_en_map"] = dataset["size"].map(data_order_2)
```
