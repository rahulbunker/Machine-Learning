```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\ASUS\OneDrive\Desktop\pandas\loan.csv")
df.head(3)
```

```python
df.isnull().sum()
```

```python
sns.distplot(df["CoapplicantIncome"])
```

```python
# IQR
# for remove the outlier
```

```python
q1 = df["CoapplicantIncome"].quantile(0.25)
q3 = df["CoapplicantIncome"].quantile(0.75)
```

```python
iqr = q3-q1
```

```python
min_r = q1-(1.5*iqr)
max_r = q3+(1.5*iqr)
```

```python
min_r,max_r
```

```python
df=df[df["CoapplicantIncome"]<=max_r]
```

```python
sns.distplot(df["CoapplicantIncome"])
```

```python
# norml distribyion ke liye ham function tranformer karenge
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft = FunctionTransformer(func=np.log1p)
ft.fit(df[["CoapplicantIncome"]])
```

```python
df["CoapplicantIncome_tf"] = ft.transform(df[["CoapplicantIncome"]])
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(df["CoapplicantIncome"])
plt.title("before")

plt.subplot(1,2,2)
sns.distplot(df["CoapplicantIncome_tf"])
plt.title("after")

plt.show()
```
