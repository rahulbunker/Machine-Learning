# Ordinal Encoding

```python
 # ordinal encoding two perfimes
# 1. sklearn
# 2. Map function
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({"size":["s","m","xl","l","s","m","xxl","s","m","l"]})
df
```

```python
data=[["s","m","l","xl","xxl"]]
```

### 1. from by Sklearn

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=data)
oe.fit(df[["size"]])
```

```python
oe.transform(df[["size"]])
```

```python
df["shape"]=oe.transform(df[["size"]])
```

```python
df
```

```python
oe=OrdinalEncoder(categories="auto")
oe.fit(df[["size"]])
# yah alphabet ke order se encoding krta hai
```

```python
oe.transform(df[["size"]])
```

```python
df["shape"]=oe.transform(df[["size"]])
```

```python
df
```

### 2. From by Map Function  

```python
df=pd.DataFrame({"size":["s","m","xl","l","s","m","xxl","s","m","l"]})
df

```

```python
new={"s":0,"m":1,"l":2,"xl":3,"xxl":5000}
```

```python
df["size"].map(new)
```

```python
df["new_size"]=df["size"].map(new)
```

```python
df
```
