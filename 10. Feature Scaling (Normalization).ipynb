```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.distplot(df["CoapplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
ms.fit(df[["CoapplicantIncome"]])
```

```python
df["CoapplicantIncome_minmax"] = ms.transform(df[["CoapplicantIncome"]])
```

```python
df.head(3)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Befor")
sns.distplot(df["CoapplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(df["CoapplicantIncome_minmax"])

plt.show()

```
