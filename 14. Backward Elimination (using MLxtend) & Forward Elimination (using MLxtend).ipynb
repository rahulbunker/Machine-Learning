```python
import pandas as pd
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
df=pd.read_csv(r"C:\Users\ASUS\OneDrive\Desktop\pandas\diabetes.csv")
df.head(3)
```

```python
x=df.iloc[:,:-1]
y=df["Outcome"]
```

```python
x.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
fs = SequentialFeatureSelector(lr,k_features=5,forward=True)
fs.fit(x,y)
```

```python
fs.feature_names
```

```python
fs.k_feature_names_
```

```python
fs.k_score_
```

```python
fs = SequentialFeatureSelector(lr,k_features=4,forward=True)
fs.fit(x,y)
fs.k_score_
```

```python
fs = SequentialFeatureSelector(lr,k_features=8,forward=True)
fs.fit(x,y)
fs.k_score_
```

```python
fs = SequentialFeatureSelector(lr,k_features=2,forward=True)
fs.fit(x,y)
fs.k_score_
```

```python
fs = SequentialFeatureSelector(lr,k_features=1,forward=True)
fs.fit(x,y)
fs.k_score_
```
