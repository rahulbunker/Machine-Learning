# Feature Scaling (Standardization)

Feature scaling is the process of adjusting the values of features in a dataset so they all fall within a similar range. This helps machine learning models perform better by ensuring no single feature dominates because of its scale.

For example:

If one feature is in kilometers and another in meters, scaling makes them comparable.
Common methods:

1. Normalization: Scales values to a range like 0 to 1.
2. Standardization: Makes values have a mean of 0 and a standard deviation of 1.

```python
# Feature Scaling 
# 1. Standardization
# 2. Normalization
```

```python
# Standardization : It is very effective technique which re-scales a feature value so
#                   that it has distribution with 0 mean value and varience equals to 1.

#   X=(Xi-Xmean)/Standard Deviation
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("C:\\Users\\ASUS\\OneDrive\\Desktop\\pandas\\loan.csv")
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
dataset["ApplicantIncome"].fillna(dataset["ApplicantIncome"].mean(),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["ApplicantIncome"])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
ss.fit(dataset[["ApplicantIncome"]])
```

```python
ss.transform(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicantIncome_ss"]=pd.DataFrame(ss.transform(dataset[["ApplicantIncome"]]),columns=["X"])
```

```python
dataset.head(5)
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("After")
sns.distplot(dataset["ApplicantIncome_ss"])
plt.subplot(1,2,2)
plt.title("Before")
sns.distplot(dataset["ApplicantIncome"])
plt.show()
```
